```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('df2.csv')
```

```python
# Calcular los valores únicos de 'customer_id' y 'churn_next_month'
churn_counts = df.groupby('customer_id')['churn_next_month'].sum().value_counts()

# Establecer el estilo de seaborn
sns.set(style="whitegrid")

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=churn_counts.index, y=churn_counts.values, palette="viridis")

# Añadir etiquetas y título
barplot.set_title("Distribución de la Suma de 'Churn Next Month' por Cliente", fontsize=16)
barplot.set_xlabel("Suma de 'Churn Next Month'", fontsize=14)
barplot.set_ylabel("Número de Clientes", fontsize=14)

# Mostrar los valores en las barras
for i in barplot.containers:
    barplot.bar_label(i, fmt='%d')

# Mostrar la gráfica
plt.show()
```

```python
type_counts = df.groupby('type')['customer_id'].nunique().reset_index()
type_counts.columns = ['type', 'unique_customer_count']

# Establecer el estilo de seaborn
sns.set(style="whitegrid")

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='type', y='unique_customer_count', data=type_counts, palette="Blues_d")

# Añadir etiquetas y título
barplot.set_title("Número de Clientes Únicos por Tipo", fontsize=16)
barplot.set_xlabel("Tipo", fontsize=14)
barplot.set_ylabel("Número de Clientes Únicos", fontsize=14)

# Inclinar las etiquetas del eje x
plt.xticks(rotation=75)

# Mostrar los valores en las barras
for i in barplot.containers:
    barplot.bar_label(i, fmt='%d')

# Mostrar la gráfica
plt.show()
```

```python
# Agrupar por 'type' y 'churn_next_month', y contar los 'customer_id' únicos
type_churn_counts = df.groupby(['type', 'churn_next_month'])['customer_id'].nunique().reset_index()
type_churn_counts.columns = ['type', 'churn_next_month', 'unique_customer_count']

# Establecer el estilo de seaborn
sns.set(style="whitegrid")

# Crear la gráfica de barras agrupadas
plt.figure(figsize=(12, 8))
barplot = sns.barplot(
    x='type', 
    y='unique_customer_count', 
    hue='churn_next_month', 
    data=type_churn_counts, 
    palette="Blues_d"
)

# Añadir etiquetas y título
barplot.set_title("Número de Clientes Únicos por Tipo y Churn", fontsize=16)
barplot.set_xlabel("Tipo", fontsize=14)
barplot.set_ylabel("Número de Clientes Únicos", fontsize=14)

# Inclinar las etiquetas del eje x
plt.xticks(rotation=75)

# Mostrar los valores en las barras
for container in barplot.containers:
    barplot.bar_label(container, fmt='%d')

# Mostrar la gráfica
plt.show()
```
